## Step 2: Calculate unweighted price indexes

### Calculate category indexes

Note: outlier defined as price change for more than 200% and less than -90%

```python
def cal_cat_index(catID):

    # read data
    data = collection.find({"catID":catID,"status":3},['date','price_change'])
    df = pd.DataFrame(list(data))
    data.close()
    
    # drop outlier
    df['price_change']=df['price_change'].fillna(0.0)
    df = df[(df['price_change']<2)&(df['price_change']>-0.9)]
    
    # calculate category index
    grouped = df.groupby(['date'])
    df_index = grouped.agg('mean')  # cat index calculatd by the mean of goods indexes
    return df_index

# calculate small category indexes
df_indexes = []
for catID in catIDs:
    df_index = cal_cat_index(catID)
    df_index['catID'] = catID
    df_index = df_index.reset_index()
    df_indexes.append(df_index)
    
df_small_index = pd.concat(df_indexes)
docs_s = df_small_index.to_dict('records')
db.tmall618_index_cat.drop()
db.tmall618_index_cat.insert_many(docs_s)
```

### Calculate Tmall indexes

```python
# category level data
try:
    df_indexes = df_small_index
except:
    data = db.tmall618_index_cat.find()
    df_indexes = pd.DataFrame(list(data))
    data.close()

# middle category index
df_indexes['catID_m']=df_indexes['catID'].apply(lambda x:str(x)[0:6])
# print(df_indexes[['catID_m','date','price_change']])
df_middle_index = df_indexes[['catID_m','date','price_change']].groupby(['catID_m','date']).agg('mean')
df_middle_index = df_middle_index.reset_index()
docs_m = df_middle_index.to_dict('records')
db.tmall618_index_m.drop()
db.tmall618_index_m.insert_many(docs_m)

# big category index
df_middle_index['catID_b']=df_middle_index['catID_m'].apply(lambda x:str(x)[0:4])
df_big_index = df_middle_index[['catID_b','date','price_change']].groupby(['catID_b','date']).agg('mean')
df_big_index = df_big_index.reset_index()
docs_b = df_big_index.to_dict('records')
db.tmall618_index_b.drop()
db.tmall618_index_b.insert_many(docs_b)

# total index
df_total_index = df_big_index[['date','price_change']].groupby(['date']).agg('mean')
df_total_index = df_total_index.reset_index()
docs_t = df_total_index.to_dict('records')
db.tmall618_index.drop()
db.tmall618_index.insert_many(docs_t)
```

2. Calculate mean frequency per category
3. Calculate median frequency and duration



####  Calculate daily frequency per individual goods

aggregate: calculate daily frequency for each goods

```python
# map-reduce
mapper = Code("""
function (){
  key = {
    "catID": this.catID,
    "goodsID": this.goodsID,
  };
  value = this.price_change;
  emit(key,value);
}
""")

reducer = Code("""
function (key,values){
  var increase = 0, decrease = 0, obs = 0;
  for (i in values){
  if (values[i]>0){
     increase = increase + 1;
     obs = obs + 1;
     }
  else if (values[i]<0){
     decrease = decrease + 1;
     obs = obs + 1;
  }
  else if (values[i]==0){
     obs = obs + 1;
     }
  else{
  
  };
  };
    return {"increase":increase/obs,"decrease":decrease/obs,"change":(increase+decrease)/obs};
};
""")

query = {
    "status":3,
    "price_change":{
        "$type":1,
    }
}

db.tmall618_freq_g.drop()
db.tmall618_freq_c.drop()
result_freq_g = collection.map_reduce(mapper, reducer, "tmall618_freq_g", query = query)
```

```python
result_freq_g.count()
```

```python
df_freq_g = pd.DataFrame(list(result_freq_g.find()))
df_freq_g = df_freq_g.dropna()
```

```python
df_freq_g['value'].head()
```

```python
df_freq_g['catID']=df_freq_g['_id'].apply(lambda x:x['catID'])
df_freq_g['goodsID']=df_freq_g['_id'].apply(lambda x:x['goodsID'])

df_freq_g['increase']=df_freq_g['value'].apply(lambda x:x['increase'])
df_freq_g['decrease']=df_freq_g['value'].apply(lambda x:x['decrease'])
df_freq_g['change']=df_freq_g['value'].apply(lambda x:x['change'])
```

### Calculate size of price changes

#### Calculate the mean size at the category level

Note: outlier defined as price change for more than 200% and less than -90%

```python
# map-reduce
mapper = Code("""
function (){
  key = this.catID;
  value = this.price_change;
  emit(key,value);
}
""")

reducer = Code("""
function (key,values){
    var size_i_sum = 0, size_i_num = 0, size_d_sum = 0, size_d_num = 0;
    for (i in values){
      if ((values[i]>0)&&(values[i]<2)){
        size_i_sum = size_i_sum + values[i];
        size_i_num = size_i_num + 1;
      }
      else if ((values[i]<0)&&(values[i]>-0.9)){
        size_d_sum = size_d_sum + values[i];
        size_d_num = size_d_num + 1;
      }
      else{
      
      };
    };
    var result = {'size_i':size_i_sum/size_i_num,'size_d':size_d_sum/size_d_num,};
    return result
}
""")

query = {
    "status":3,
    "price_change":{
        "$type":1,
    },
}


db.tmall618_size_c.drop()
result_size_c = collection.map_reduce(mapper, reducer, "tmall618_size_c", query = query)
```

```python
df_size_c = pd.DataFrame(list(result_size_c.find()))
df_size_c = df_size_c.dropna()
df_size_c.columns
```
